# IF_ELSE

```python
#CONTOH PENGGUNAAN BOOLEANS PADA METHOD IF ELSE
a = int(input("Masukkan Angka pertama = "))
b = int(input("Masukkan Angka kedua = "))
if a > b:
  print("Angka pertama lebih besar dari Angka kedua")
elif a < b:
  print("Angka kedua lebih besar dari Angka Pertama")
else:
    print("Angka pertama sama dengan Angka kedua")
```

```python
#CONTOH IF ELSE YANG MENGGUNAKAN AND, OR, NOT.

# A N D
"""
TRUE = T -- FALSE = F
T + T = T
T + F = F
F + T = F
F + F = F
JADI, JIKA KITA TARIK KESIMPULAN >> FUNGSI AND AKAN TRUE JIKA SEMUA KONDISINYA TERPENUHI
"""
#CONTOH SEDERHANA
a = int(input("Masukkan Angka pertama = "))
b = int(input("Masukkan Angka kedua = "))
if a > b and a > 100:
  print("Angka pertama lebih besar dari Angka kedua dan Lebih besar dari 100")
elif a > b and a < 100:
  print("Angka pertama lebih besar dari Angka kedua tetapi Lebih kecil dari 100")
elif a < b and b > 100:
  print("Angka kedua lebih besar dari Angka Pertama dan Lebih besar dari 100")
elif a < b and b < 100:
  print("Angka kedua lebih besar dari Angka Pertama tetapi Lebih kecil dari 100")
else:
 print("Angka pertama sama dengan Angka kedua")
```

```python
# O R
"""
TRUE = T -- FALSE = F
T + T = T
T + F = T
F + T = T
F + F = F
JADI, JIKA KITA TARIK KESIMPULAN >> FUNGSI AND AKAN TRUE JIKA SALAH SATU KONDISINYA TERPENUHI
"""
#CONTOH SEDERHANA
a = int(input("Masukkan Angka pertama = "))
b = int(input("Masukkan Angka kedua = "))
if a > b or a == b:
 print("Angka pertama >= Angka kedua")
elif a < b or a == b:
 print("Angka kedua >= Angka pertama")
else:
 print("Angka pertama sama dengan Angka kedua")
```

```python
# N O T
"""
TRUE = T -- FALSE = F
T = F
F = T
JADI, JIKA KITA TARIK KESIMPULAN >> FUNGSI AND AKAN TRUE JIKA SALAH SATU KONDISINYA TERPENUHI
"""
#CONTOH SEDERHANA
a = int(input("Masukkan Angka pertama = "))
b = int(input("Masukkan Angka kedua = "))
if not(a > b):
  print("Angka pertama lebih kecil dari Angka kedua")
elif not(a < b):
  print("Angka kedua lebih kecil dari Angka Pertama")
else:
 print("Angka pertama sama dengan Angka kedua")
```

# TUPLES

```python
#UNTUK FUNGSI TUPLES TERSENDIRI HAMPIR MIRIP DENGAN FUNGSI LIST, MUNGKIN YANG MEMBEDAKANNYA ADALAH JIKA INGIN UPDATE TUPLES HARUS MENGCONVERTNYA TERLEBIH DAHULU MENJADI LIST
#CONTOH SEDERHANA
fruits = ("apple", "banana", "cherry", "manggo", "kiwi")
print(fruits)
print(len(fruits)) # >> BERFUNGSI MENUNJUKKAN BANYAK ITEM YANG TERDAFTAR
type(fruits)
```

```python
#CARA MENAMBAHKAN ITEM
fruits.append("orange")
print(fruits)

#ATAU KITA JUGA BISA MENAMBAHKAN ITEM TETAPI DENGAN BERBEDA JENIS DATA SEPERTI STR DENGAN INT
fruits.append(100)
print(fruits)

#PROSES MEMANGGIL INDEX
fruits = ("apple", "banana", "cherry", "manggo", "kiwi")
print(fruits[1])
print(fruits[-1])  #PROSES PEMANGGILAN INDEX TETAPI DIHITUNG DARI INDEX PALING AKHIR
print(fruits[2:5]) #CARA MEMANGGIL MULTI INDEX
```

```python
#UPDATE TUPLES
fruits = ("apple", "banana", "cherry", "manggo", "kiwi")
buah = list(fruits)
buah[1] = "kiwi"
fruits = tuple(buah)
print(fruits)
```

```python
#UNPACK TUPLES
fruits = ("apple", "banana", "cherry")
(green, yellow, red) = fruits
print(green)
print(yellow)
print(red)
```

# DICTIONORIES

```python
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
print(cars)
```

```python
#GET ITEM IN DICTIONORIES
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = cars["model"]
print(x)

cars =	{
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
x = cars.get("model")
print(x)
```

```python
#EDIT / CHANGE ITEM IN DICTIONORIES
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars["year"] = 2018
print(cars)

cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars.update({"year": 2020})
print(cars)
```

```python
#ADD ITEMS IN DICTIONORIES
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars["color"] = "red"
print(cars)

cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars.update({"color": "blue"})
print(cars)
```

```python
#REMOVE ITEMS IN DICTIONORIES
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars.pop("model")
print(cars)


#MENGHAPUS SEMUA ITEM YANG ADA DI DICTIONORIES
cars = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964
}
cars.clear()
print(cars)
```
